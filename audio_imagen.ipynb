```python
#%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd #audio
from imageio import imread #scipy.misc.imread fue/ser√° eliminado!
```

# Audio Generado

```python
sr = 22050 # Tasa de sampleo
f = 440. #Hz
T = 1/f # periodo en segundos
tmax = 2. # tiempo max en segundos
t = np.linspace(0, tmax, int(tmax*sr), endpoint=False) # tiempo variable
A = 0.5 # amplitud de la onda
x = A*np.sin(2*np.pi*f*t)# onda sinusoidal a f=440 Hz
```

```python
T
```

```python
x
```

```python
t
```

```python
ipd.Audio(x, rate=sr) #reprodude el array1d de numpy
```

```python
plt.plot(t,x,'.')
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
```

```python
ipd.Audio(x[0:200], rate=sr) #reprodude el array1d de numpy
```

```python
plt.plot(t[0:200],x[0:200],'.')
plt.xlabel('$t$')
plt.ylabel('$x(t)$')
```
